```python
from pysearchlib.share.ExpAnaConf import ANAConf
from pysearchlib.share.ExcelOutKeys import PaperFormat
from matplotlib import pyplot as plt, ticker
import seaborn as sns

fig, axes = plt.subplots(3, 3, figsize=(12, 8))
plt.subplots_adjust(wspace=0.1, hspace=0.52)
import matplotlib as mpl

mpl.rcParams['axes.linewidth'] = 0.5
mpl.rcParams['axes.linewidth'] = 0.4
sns.set_theme(style="ticks")
TIME_SCALE = 1.0 / (60 * 60)
for _row_index, _opt_metri in enumerate([EK.VUS_ROC, EK.VUS_PR, EK.RF]):
    _data = OverallStatics(target_perf=_opt_metri).load_data_effect_of_sample_method_and_interval()
    for _col_index, (_sample_method, _values) in enumerate(_data.groupby(by="抽样方法")):
        _filter = _values
        ax = axes[_row_index, _col_index]
        ax.grid(False)
        ax2 = ax.twinx()
        _title=f"""$\mathbf{{({ANAConf.convert_number_to_alpha(_row_index * 3 + _col_index + 1)})}}$ {PaperFormat.format_value(_opt_metri)} and  $T_{{fast}}$ w.r.t. {PaperFormat.format_value(_sample_method)} """
        ax.set_title(_title,y=-0.5)
        # if _row_index<=1:
        #     ax.set_title(_title,y=-0.4)
        # else:
        #     ax.set_title(_title,y=-0.5)

        ax2.set_ylabel("Training time (hours)")
        ax.set_xlabel("Sampling gap")
        ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))
        # ax2.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))
        ax2.yaxis.set_major_locator(ticker.MultipleLocator(5))
        ax.set_ylabel("Accuracy")
        ax2.set_ylim(5, 60)
        if _opt_metri == EK.VUS_ROC:
            # ax.set('set y2r [68.0:73.5]')
            ax.set_ylim(68, 73.6)
            ax.yaxis.set_major_locator(ticker.MultipleLocator(.75))
            # ax.set_ylabel("VUS ROC")
        if _opt_metri == EK.VUS_PR:
            # gp.set('set y2r [40.0:42.1]')
            ax.set_ylim(40, 42.1)
            ax.yaxis.set_major_locator(ticker.MultipleLocator(.25))
            # ax.set_ylabel("VUS PR")
        if _opt_metri == EK.RF:
            ax.set_ylim(21.5, 24.1)
            ax.yaxis.set_major_locator(ticker.MultipleLocator(.5))
            # ax.set_ylabel("RF")
        sns.pointplot(_filter,
                      x=_filter['抽样间隔'],
                      y=_filter["fastuts auc. avg."],
                      marker="+",
                      errorbar=None,
                      linewidth=1,
                      markersize=10,
                      markeredgewidth=1.5, color=COLOR_1,
                      label="Accuracy",
                      ax=ax)
        sns.pointplot(x=_filter['抽样间隔'],
                      y=_filter["FastUTS 训练时间(sec)"] * TIME_SCALE, marker="x", errorbar=None, linewidth=1,
                      markersize=10,
                      markeredgewidth=1.5,
                      label="Training Time",
                      color=COLOR_2, ax=ax2)
        if _col_index != 0:
            ax.set_ylabel(None)
            ax.get_yaxis().set_visible(False)
        if _col_index != 2:
            ax2.set_ylabel(None)
            # ax2.tick_params(axis='y', which='both', labelleft=False, left=False)
            ax2.get_yaxis().set_visible(False)
            # ax2.tick_params(axis='y', which='both', labelleft=False, left=False)
        # if _row_index != 2:
        #     ax.set_xlabel(None)

        ax.legend().set_visible(False)
        ax2.legend().set_visible(False)

lines, labels = fig.axes[-1].get_legend_handles_labels()
lines1, labels1 = fig.axes[1].get_legend_handles_labels()
# bbox_to_anchor=(0.1, -0.1)
fig.legend(lines + lines1, labels + labels1, loc='upper center', ncol=3, frameon=False, bbox_to_anchor=(0.5, 0.94))

plt.savefig("Fig.6.pdf", bbox_inches='tight')
```
